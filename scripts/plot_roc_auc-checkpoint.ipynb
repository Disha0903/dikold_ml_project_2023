```python
##Imports
import pandas as pd
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt

sp20k = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/spBlock20.csv')
sp200k = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/spBlock200.csv')
sp400k = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/spBlock400.csv')
sp700k = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/spBlock700.csv')
random = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/random_pred.csv')
buf = pd.read_csv('C:/Users/User/Desktop/project_ml_team29/results/buff_cur_pred.csv')
```

```python
sp200k = sp20k.rename({'occurrenceStatus.1': 'occurrenceStatus', 'pred.1': 'pred'}, axis=1) 
sp400k = sp400k.rename({'occurrenceStatus.2': 'occurrenceStatus', 'pred.2': 'pred'}, axis=1) 
sp700k = sp700k.rename({'occurrenceStatus.3': 'occurrenceStatus', 'pred.3': 'pred'}, axis=1) 
buf = buf.rename({'occurrenceStatus': 'geometry.2'}, axis=1) 
buf = buf.rename({'Unnamed: 0' : 'occurrenceStatus'}, axis=1)
```

```python
##Prepare data
sp200k = sp20k.rename({'occurrenceStatus.1': 'occurrenceStatus', 'pred.1': 'pred'}, axis=1) 
sp400k = sp400k.rename({'occurrenceStatus.2': 'occurrenceStatus', 'pred.2': 'pred'}, axis=1) 
sp700k = sp700k.rename({'occurrenceStatus.3': 'occurrenceStatus', 'pred.3': 'pred'}, axis=1) 
buf = buf.rename({'occurrenceStatus': 'geometry.2'}, axis=1) 
buf = buf.rename({'Unnamed: 0' : 'occurrenceStatus'}, axis=1)

sp20k['pred'] = sp20k['pred'].apply(lambda x: float(x.split()[0].replace(',', '.')))
sp200k['pred'] = sp200k['pred'].apply(lambda x: float(x.split()[0].replace(',', '.')))
sp400k['pred'] = sp400k['pred'].apply(lambda x: float(x.split()[0].replace(',', '.')))
sp700k['pred'] = sp700k['pred'].apply(lambda x: float(x.split()[0].replace(',', '.')))
```

```python
plt.figure(0).clf()
fpr, tpr, thresholds = metrics.roc_curve(sp20k['occurrenceStatus'], sp20k['pred'])
# Calculate Area under the curve to display on the plot
auc = round(metrics.roc_auc_score(sp20k['occurrenceStatus'], sp20k['pred']),3)
# Now, plot the computed values
plt.plot(fpr, tpr, label="rf spBlock20km, AUC="+str(auc))
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('1-Specificity(False Positive Rate)')
plt.ylabel('Sensitivity(True Positive Rate)')
plt.title('Test ROC AUC')



fpr, tpr, thresholds = metrics.roc_curve(sp200k['occurrenceStatus'], sp200k['pred'])
auc = round(metrics.roc_auc_score(sp200k['occurrenceStatus'], sp200k['pred']),3)
plt.plot(fpr, tpr, label="rf spBlock200km, AUC="+str(auc))


fpr, tpr, thresholds = metrics.roc_curve(sp400k['occurrenceStatus'], sp400k['pred'])
auc = round(metrics.roc_auc_score(sp400k['occurrenceStatus'], sp400k['pred']),3)
plt.plot(fpr, tpr, label="rf spBlock400km, AUC="+str(auc))


fpr, tpr, thresholds = metrics.roc_curve(sp700k['occurrenceStatus'], sp700k['pred'])
auc = round(metrics.roc_auc_score(sp700k['occurrenceStatus'], sp700k['pred']),3)
plt.plot(fpr, tpr, label="rf spBlock700km, AUC="+str(auc))

fpr, tpr, thresholds = metrics.roc_curve(random['occurrenceStatus'], random['pred'])
auc = round(metrics.roc_auc_score(random['occurrenceStatus'], random['pred']),3)
plt.plot(fpr, tpr, label="rf randomBlock, AUC="+str(auc))

fpr, tpr, thresholds = metrics.roc_curve(buf['occurrenceStatus'], buf['pred'])
auc = round(metrics.roc_auc_score(buf['occurrenceStatus'], buf['pred']),3)
plt.plot(fpr, tpr, label="rf B-LOO, AUC="+str(auc))

plt.legend()
```
